```python
import random

# seed input
seed = int(input('Enter ANY integer number from your head: '))
print('You entered as seed for the random number generator:', seed)

# seed to initialize random number generator
random.seed(seed)

# the initial velocity of the cannonball will be between 50 m/s and 100 m/s
v0_min = 50
v0_max = 100

v0 = random.randint(v0_min, v0_max)
print('The initial velocity of the cannonball is:', v0, 'm/s')

# the velocity of the space invader will be between 10 m/s and 15 m/s
vs_min = 10
vs_max = 15

vs = random.randint(vs_min, vs_max)
print('The velocity of the space invader is:', vs, 'm/s')
```
