```python
import random

# seed input
seed = int(input('Enter ANY integer number of your choice: '))
print('You entered as seed for the random number generator:', seed)

# seed to initialize random number generator
random.seed(seed)

# the inductance of the inductor in the circuit will be between 50 miliHenry and 100 miliHenry
L_min = 50
L_max = 100

inductance = random.randint(L_min, L_max)
print('The inductance is:', inductance, 'miliHenry (mH)')

# the resistance of the resistor in the circuit will be between 5 Ohms and 20 Ohms
R_min = 5
R_max = 20

resistance = random.randint(R_min, R_max)
print('The resistance is:', resistance, 'Ohm')

# the capacitance of the capacitor in the circuit will be between 250 microFarad and 350 microFarad
C_min = 250
C_max = 350

capacitance = random.randint(C_min, C_max)
print('The capacitance is:', capacitance, 'microFarad')
```
