```python
import matplotlib.pyplot as plt
import numpy as np
# Code taken from matplotlib manual at 
# https://matplotlib.org/stable/gallery/lines_bars_and_markers/simple_plot.html

# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='About as simple as it gets, folks')
ax.grid()

# if one wishes to save the plot onto a file, use the instruction below.
fig.savefig("test.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,4*np.pi,0.1)

y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y1, label='sin(x)')
plt.plot(x,y2, label='cos(x)')

plt.xlabel('x')
plt.ylabel('y')

plt.title('Trigonometric functions')
plt.legend(loc='lower right')
plt.savefig('sin_cos_plot.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# This example is for a figure with 2x2 subplots
fig, panel = plt.subplots(2,2, constrained_layout=True)

# creating data points using numpy arrays
x = np.arange(1,10)

# plotting numerous functions in subplot mode
panel[0,0].plot(x,x*x)
panel[0,0].set_title('(a) square')
panel[0,1].plot(x,np.sqrt(x))
panel[0,1].set_title('(b) square root')
panel[1,0].plot(x,np.exp(x))
panel[1,0].set_title('(c) exp')
panel[1,1].plot(x,np.log10(x))
panel[1,1].set_title('(d) log')

# This loop adds axis labels in the horizontal and vertical axes
for ax in panel.flat:
    ax.set(xlabel='x', ylabel='y')
    
# If all axis labels are common, one can use the for loop below
# to hide x labels and tick labels for top plots and y ticks for the right-hand side plots.
# Uncomment the lines below to see how it works.
#for ax in panel.flat:
#    ax.label_outer()    
    
plt.show()
```
