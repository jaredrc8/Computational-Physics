This code generates the initial conditions for the metallic bar problem in exercise 4 (Finite difference method) of our final exam. Do not alter this code. Simply run the code as instructed and take the temperature settings to be used in your simulation for the exercise.

```python
import random

# seed input
seed = int(input('Enter ANY integer number from your head: '))
print('You entered as seed for the random number generator:', seed)

# seed to initialize random number generator
random.seed(seed)

# the temperature (in degree Celsius) on the left extreme of the bar (x=0) will be between 0 and 20
Tl_min = 0.0
Tl_max = 20.0

T_left = random.uniform(Tl_min, Tl_max)
print('The temperature (in degree Celsius) on the left extreme of the bar (x=0) is:', T_left)

# the temperature (in degree Celsius) on the right extreme of the bar (x=L) will be between 80 and 110
Tr_min = 80.0
Tr_max = 110.0

T_right = random.uniform(Tr_min, Tr_max)
print('The temperature (in degree Celsius) on the right extreme of the bar (x=L) is:', T_right)

# the length of the metallic bar (in meters)
L_min = 4
L_max = 10

L = random.randint(L_min, L_max)
print('The total length of the metallic bar (in meters) is:', L)

```
