```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import solve_ivp

#dy/dt = -y/2
def fun(t,y): return -0.5*y

# limits of the time interval
time_interval = [0,10]
y0 = [2]

# time interval filled with points
t_eval = np.arange(time_interval[0], time_interval[1], 0.1)

# ODE solver
solution = solve_ivp(fun, time_interval, y0, t_eval = t_eval)

exact_sol = 2.0*np.exp(-t_eval/2.0)
#print(solution)

print('time:')
print(solution.t)

print('y(t):')
print(solution.y)

plt.figure()
plt.plot(solution.t, solution.y[0],'o')
plt.plot(solution.t, exact_sol)
plt.show()
```
